```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Titanic_visu.csv')
```

```python
embarked_counts = df['Embarked'].value_counts()
embarked_counts.plot(kind='bar')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.title('Passenger Embarkation Counts')
plt.show()
```

```python
import numpy as np
N = 15
plt.figure(figsize = (15, 15))
#see the spelling of arrange
ind = np.arange(N)
width = 0.25 
xval = df['Fare'].head(15)

bar1 = plt.bar(ind, xval, width, color = 'r')
yval = df['New_Fare'].head(15)
bar2 = plt.bar(ind + width, yval, width, color = 'b')

#comparitive charts should always have legends

plt.xlabel('Name')
plt.ylabel('Fare')
plt.xticks(ind + width, df['Name'].head(15))
plt.legend((bar1, bar2),('Fare', 'New_Fare'))
```

```python
plt.barh(df['Embarked'],df['Fare'] )
plt.title("Embarked vs Fare")
plt.xlabel("Embarked")
plt.ylabel("Fare")
```

```python
plt.figure(figsize=(9,5))
plt.bar(df['Fare'].head(10), df['Fare'].head(10))
plt.bar(df['Fare'].head(10), df['New_Fare'].head(10), bottom = df['Fare'].head(10))
```
